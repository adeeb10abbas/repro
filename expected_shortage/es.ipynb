```python
## Let's plot all the lower bounds and all the ESes in a meaningful way. 
import numpy as np
import matplotlib.pyplot as plt

# Setting a seed for reproducibility
np.random.seed(0)

# Generate random data to represent outcomes (number of tasks completed per day)
outcomes = np.random.randint(5, 12, size=10)  # Random integers between 5 and 12

# Initialize lists to store the data
lower_bounds = range(5, 15)  # Extended and more realistic range of lower bounds
expected_shortages = []

# Calculate the shortages for each lower bound and store the expected shortages
for lower_bound in lower_bounds:
    shortages = np.maximum(lower_bound - outcomes, 0)
    expected_shortage = np.mean(shortages)
    expected_shortages.append(expected_shortage)
    print("Lower Bound: ", lower_bound, "Expected Shortage: ", expected_shortage)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(lower_bounds, expected_shortages, marker='o', linestyle='-', color='b')
plt.title('Expected Shortage vs. Lower Bound')
plt.xlabel('Lower Bound (Number of Tasks)')
plt.ylabel('Expected Shortage (Tasks)')
plt.grid(True)
plt.show()
```

## Maximum Expected Shortage and Optimality Criteria (Uniformly Most Accurate: UMA)

- it's hard to compute MES because if it's not concave we can get stuck in local maximas, which makes total sense. There's some more it. But I will just move on for now 
- The paper then moves back to UMA to get a lower bound for an unknown success probability. 
- a basic UMA example below - 





```python
import numpy as np
from scipy.stats import norm

# Generate sample data
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=100)  # mean=50, std=10, n=100

# Significance level
alpha = 0.05

# Calculate sample mean and standard error
sample_mean = np.mean(data)
se = np.std(data, ddof=1) / np.sqrt(len(data))

# Calculate the UMA lower bound
z_score = norm.ppf(alpha)
uma_lower_bound = sample_mean + z_score * se

print(f"Sample Mean: {sample_mean:.2f}")
print(f"UMA Lower Confidence Bound: {uma_lower_bound:.2f}")
```
